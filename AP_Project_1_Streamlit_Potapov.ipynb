```python
import pandas as pd
import numpy as np
np.random.seed(42)  # Добавим, чтобы одинаковые данные генерировались

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

### Аномалии

Выявить аномалии, где температура выходит за пределы $ \text{среднее} \pm 2\sigma $.

```python
# Аномалии за пределами среднего +- 2std по всем записям
lower_border = df['temperature'].mean() - 2 * df['temperature'].std()
upper_border = df['temperature'].mean() + 2 * df['temperature'].std()
df_with_anomalies = df.query("temperature < @lower_border or temperature > @upper_border")

print(f"Аномалии за пределами среднего +- 2std по всем записям")
df_with_anomalies
```

```python
# Аномалии внутри каждого города и сезона
df_mean_each_season['lower_border'] = df_mean_each_season['mean'] - 2 * df_mean_each_season['std']
df_mean_each_season['upper_border'] = df_mean_each_season['mean'] + 2 * df_mean_each_season['std']

df_merged = df.merge(df_mean_each_season, on=['city', 'season'])
df_with_anomalies = df_merged.query("temperature < lower_border or temperature > upper_border")

print(f"Аномалии внутри каждого города и сезона")
df_with_anomalies
```

### Распараллеливание анализа

Попробуйте распараллелить проведение этого анализа. Сравните скорость выполнения анализа с распараллеливанием и без него.

```python
from joblib import Parallel, delayed

# Функция анализа одного города
def analyze_city(df_city):

    # Считаем границы
    df_city_season_stats = df_city.groupby('season')['temperature'].agg(['mean', 'std']).reset_index()
    df_city_season_stats['lower_border'] = df_city_season_stats['mean'] - 2 * df_city_season_stats['std']
    df_city_season_stats['upper_border'] = df_city_season_stats['mean'] + 2 * df_city_season_stats['std']

    # Добавляем к датафрейму
    df_city = df_city.merge(df_city_season_stats, on='season')
    
    # Аномалии
    df_city['anomaly_flg'] = (df_city['temperature'] < df_city['lower_border']) | (df_city['temperature'] > df_city['upper_border'])
    
    return df_city

# Распараллеливание по городам
list_cities = list(df['city'].unique())
results = Parallel(n_jobs=-1)(delayed(analyze_city)(df[df['city'] == city]) for city in list_cities)

# Объединяем результаты
df_result = pd.concat(results).reset_index(drop=True)
df_result.query("anomaly_flg")
```

Результат получили тот же, сравним время анализа

```python
import time

def analyze_cities_no_parallel(df):

    # Считаем границы
    df_mean_each_season = df.groupby(['city', 'season'])['temperature'].agg(['mean', 'std']).reset_index()
    df_mean_each_season['lower_border'] = df_mean_each_season['mean'] - 2 * df_mean_each_season['std']
    df_mean_each_season['upper_border'] = df_mean_each_season['mean'] + 2 * df_mean_each_season['std']

    # Добавляем к датафрейму
    df = df.merge(df_mean_each_season, on=['city', 'season'])
    
    # Аномалии
    df['anomaly_flg'] = (df['temperature'] < df['lower_border']) | (df['temperature'] > df['upper_border'])
    
    return df

df_result = analyze_cities_no_parallel(df)
df_result.query("anomaly_flg")
```

### Нормальность текущей температуры

Определить, является ли текущая температура нормальной, исходя из исторических данных для текущего сезона.

```python
import datetime

# Функции
def get_response_season(response_dt):

    response_month = datetime.datetime.fromtimestamp(response_dt, tz=datetime.timezone.utc).month

    dict_month_x_season = {1: 'winter', 2: 'winter', 12: 'winter',
                           3: 'spring', 4: 'spring', 5: 'spring',
                           6: 'summer', 7: 'summer', 8: 'summer',
                           9: 'autumn', 10: 'autumn', 11: 'autumn'}
    
    return dict_month_x_season[response_month]


def get_current_weather(api_key, city):

    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        curent_temperature = data['main']['temp']
        curent_season = get_response_season(data['dt'])
        return curent_temperature, curent_season
    else:
        raise Exception(response.text)


def check_normality_of_temperature(api_key, city, df):

    # Плучаем температуру и сезон
    curent_temperature, curent_season = get_current_weather(api_key=api_key, city=city)

    # Фильтруем
    df_filtered = df.query("city == @city and season == @curent_season")

    # Получаем границы
    lower_border = df_filtered['temperature'].mean() - 2 * df_filtered['temperature'].std()
    upper_border = df_filtered['temperature'].mean() + 2 * df_filtered['temperature'].std()

    if lower_border <= curent_temperature <= upper_border:
        print(f"Текущая температура в городе {city} составляет {curent_temperature} и является нормальной для сезона {curent_season}. Границы нормальности: {round(lower_border, 2)} и {round(upper_border, 2)}")
    else:
        print(f"Текущая температура в городе {city} составляет {curent_temperature} и НЕ является нормальной для сезона {curent_season}. Границы нормальности: {round(lower_border, 2)} и {round(upper_border, 2)}")
    
    return

check_normality_of_temperature(api_key=API_KEY, city='Moscow', df=df)
```

### Асинхронность

Попробуйте для получения текущей температуры использовать синхронные и асинхронные методы. Что здесь лучше использовать?

```python
list_cities = list(df['city'].unique())
```

```python
# Время синхронной обработки запросов
start_time = time.time()
for city in list_cities:
    check_normality_of_temperature(api_key=API_KEY, city=city, df=df)
end_time = time.time()

print(f"\nВремя синхронной обработки запросов: {round(end_time - start_time, 4)} секунд")
```

```python
import aiohttp
import asyncio

# Функции для асинхронной обработки
async def async_check_normality_of_temperature(api_key, city, df, session):

    url = f"http://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"

    async with session.get(url) as response:

        # Получаем температуру и сезон
        data = await response.json()
        current_temerature = data['main']['temp']
        current_season = get_response_season(data['dt'])
        
        # Фильтруем исторические данные
        df_filtered = df.query("city == @city and season == @current_season")
        lower_border = df_filtered['temperature'].mean() - 2 * df_filtered['temperature'].std()
        upper_border = df_filtered['temperature'].mean() + 2 * df_filtered['temperature'].std()
        
        normal_flg = lower_border <= current_temerature <= upper_border
        status = "НОРМАЛЬНАЯ" if normal_flg else "НЕ является нормальной"
        
        return (city, current_temerature, current_season, lower_border, upper_border, normal_flg)


async def async_check_all_cities(list_cities, api_key, df):

    async with aiohttp.ClientSession() as session:
        tasks = [async_check_normality_of_temperature(api_key, city, df, session) for city in list_cities]
        results = await asyncio.gather(*tasks)

    return results
```

```python
# Время асинхронной обработки
start_time = time.time()
results = await async_check_all_cities(list_cities=list_cities, api_key=API_KEY, df=df)
for res in results:
    print(res)
end_time = time.time()

print(f"\nВремя Асинхронной обработки запросов: {round(end_time - start_time, 4)} секунд")
```

**Асинхронность дала большое увеличение скорости, примерно в 10-13 раз**
